```python
import numpy as np
import matplotlib.pyplot as plt

fgsmeps=np.array([0.15,0.3,0.45,0.6])
jsmatheta=np.array([0.5,0.75,1.0,1.25])
baiters=np.array([1000,2000,3000,4000])
cwiters=np.array([10,50,80,100])

fgsmacc=np.array([0.6620,0.1090,0.0290,0.0260])
fgsmsc=np.array([0.00461915,0.00922415,0.0139298,0.0184118])
fgsmt=np.array([0.3561553955078125,0.2935309410095215,0.3624598979949951,0.318805456161499])/1000

jsmaacc=np.array([0.2610,0.0990,0.0450,0.0210])
jsmasc=np.array([0.00432502,0.00511915,0.00586339,0.00581703])
jsmat=np.array([690.9513363838196,498.86871457099915,408.20263266563416,371.8326008319855])/50

baacc=np.array([0,1,9,23])/50
basc=np.array([0.00326445288196,0.00282748073948,0.002547927809,0.00285387738627])
bat=np.array([724.2450070381165,1203.1902935504913,1694.1352043151855,2122.0194647312164])/50

cwacc=np.array([0.4,1,1,1])
cwsc=np.array([1.2e-3,4.2e-3,3.1e-3,2.7e-3])
cwt=np.array([0.612,3.316,4.122,5.328])
```

```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

```python
plt.cla()
plt.clf()
plt.figure(figsize=(4,4))
plt.plot(fgsmeps,fgsmacc,'r-',label="Accuracy")
plt.plot(fgsmeps,fgsmsc*100,'b-',label=r"Score $(\times 10^{-3})$")
plt.plot(fgsmeps,fgsmt*1000,'y-',label=r"Time ($10^{-3}$ seconds)")
plt.title(r"Performance of FGSM vs. $\epsilon$")
plt.xlabel(r"$\epsilon$")
plt.legend()
plt.show()
```

```python
plt.cla()
plt.clf()
plt.figure(figsize=(4,4))
plt.plot(jsmatheta,jsmaacc,'r-',label="Accuracy")
plt.plot(jsmatheta,jsmasc*100,'b-',label=r"Score $(\times 10^{-3})$")
plt.plot(jsmatheta,jsmat/10,'y-',label=r"Time ($10$ seconds)")
plt.title(r"Performance of JSMA vs. $\theta$")
plt.xlabel(r"$\theta$")
plt.legend()
plt.show()
```

```python
plt.cla()
plt.clf()
plt.figure(figsize=(4,4))
plt.plot(baiters,baacc,'r-',label="Accuracy")
plt.plot(baiters,basc*100,'b-',label=r"Score $(\times 10^{-3})$")
plt.plot(baiters,bat/100,'y-',label=r"Time ($100$ seconds)")
plt.title(r"Performance of Boundary Attack vs. Iterations")
plt.xlabel(r"Maximum Iterations")
plt.legend()
plt.show()
```

```python
plt.cla()
plt.clf()
plt.figure(figsize=(4,4))
plt.plot(cwiters,cwacc,'r-',label="Accuracy")
plt.plot(cwiters,cwsc*100,'b-',label=r"Score $(\times 10^{-3})$")
plt.plot(cwiters,cwt/10,'y-',label=r"Time ($10$ seconds)")
plt.title(r"Performance of Boundary Attack vs. Iterations")
plt.xlabel(r"Maximum Iterations")
plt.legend()
plt.show()
```
